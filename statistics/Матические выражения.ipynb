3. **Целые числа** &mdash; включают в себя натуральные числа, числа противоположные натуральным (т.е. с отрицательным знаком) и ноль.
    $\mathbb{Z = Z^-\ \cup\ }\{0\}\mathbb{\ \cup\  Z^+} = \{...,-3,-2,-1,0,1,2,3,...\}$

    **Целые положительные числа:** $\mathbb{Z^+} = \mathbb{N} = \{1, 2, 3, ...\}$

    **Целые отрицательные числа:**
    $\mathbb{Z^{}_-} = \{..., -3, -2, -1 \}$
